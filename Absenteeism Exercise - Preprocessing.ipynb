## Determining variables of interest

```python
df = df.drop(['ID'], axis = 1)
df.describe(include='all')
```

## Creating the dummy variables

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns.head()
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns.head()
```

```python
reason_columns['check'].sum(axis=0)
```

```python
reason_columns['check'].unique()
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first = True)
```

```python
reason_columns
```

```python
reason_columns.columns.values
```

```python
reason_columns.loc[:, '1': '14'].max(axis=1)
```

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

### Concatenate Converted Column Values

```python
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
df.head()
```

```python
df.columns.values
```

```python
column_names = ['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
df.columns = column_names
```

```python
df.head()
```

## Reorder Columns

```python
column_names_reordered = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']
```

```python
df = df[column_names_reordered]
df.head()
```

## Checkpoint

```python
df_reason_mod = df.copy()
```

## Fixing 'Date'

```python
type(df_reason_mod['Date'][0])
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
df_reason_mod.head()
```

```python
df_reason_mod.info()
```

### Extract the Month value

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []

for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)

df_reason_mod['Month Value'] = list_months
df_reason_mod.head()
```

### Extracting the day of the week

```python
df_reason_mod['Date'][0].weekday()
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the Week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

```python
df_reason_mod.drop(['Date'], axis = 1)
```

```python
df_date_reason_mod = df_reason_mod.drop(['Date'], axis=1)
```
